```python
from scipy.signal import savgol_filter
```

```python
window_size = 9 
poly_order = 3

# Apply the Savitzky-Golay filter
smoothed_extent = savgol_filter(Bell_Amundsen_Extent['Extent'].values, window_size, poly_order)

# Create a new DataFrame with smoothed values
df_smoothed = pd.DataFrame({'Date': Bell_Amundsen_Extent['Date'], 'Extent': smoothed_extent})
```

```python
plt.figure(figsize=(16, 8))
plt.plot(Bell_Amundsen_Extent['Date'], Bell_Amundsen_Extent['Extent'], label='Original Data', alpha=0.7)
plt.plot(Bell_Amundsen_Extent['Date'], smoothed_extent, label=f'Savitzky-Golay Smoothed (window_size={window_size}, poly_order={poly_order})', color='red')
plt.title('Savitzky-Golay Smoothing of Ice Extent Data')
plt.xlabel('Date')
plt.ylabel('Ice Extent (million sq km)')
plt.legend()
plt.show()
```

```python
fp.plot_df(df_smoothed)
```

```python
scaler = MinMaxScaler()
df_smoothed['Extent'] = scaler.fit_transform(df_smoothed['Extent'].values.reshape(-1, 1))
```

```python
extent = df_smoothed['Extent']
W_S=5
X,y = fp.df_to_X_y(extent, W_S)
X.shape, y.shape
```

```python
X, y = shuffle(X, y, random_state=42)
```

```python
def create_lstm():
    model = Sequential()
    model.add(InputLayer((W_S, 1)))
    model.add(LSTM(64, activation='relu', kernel_constraint=MaxNorm(1.0), return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(32, activation='relu', kernel_constraint=MaxNorm(1.0)))
    model.add(Dense(1, activation='linear'))
    model.compile(loss=MeanSquaredError(), optimizer=Adam(), metrics=[RootMeanSquaredError()])
    return model
```

```python
def k_fold_cross_validation(X, y, scaler, k=5):
    tscv = TimeSeriesSplit(n_splits=k)
    mse_scores = []

    for i, (train_index, test_index) in enumerate(tscv.split(X)):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Create ModelCheckpoint callback
        checkpoint_filepath = f'Bell_Amundsen_model_checkpoint_fold_{i + 1}.h5'
        model_checkpoint = ModelCheckpoint(checkpoint_filepath, save_best_only=True, monitor='val_loss', mode='min', verbose=1)

        # Create and train GRU model
        model = create_lstm()
        history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=12, callbacks=[model_checkpoint])

        # Load the best model based on validation loss
        model.load_weights(checkpoint_filepath)

        # Plot training and validation loss
        plt.plot(history.history['loss'], label=f'Training Loss - Fold {i + 1}')
        plt.plot(history.history['val_loss'], label=f'Validation Loss - Fold {i + 1}')

        # Make predictions on the test set
        y_pred_scaled = model.predict(X_test)
        y_test_scaled = y_test.reshape(-1, 1)

        # Inverse transform predictions and original test data
        y_pred = scaler.inverse_transform(y_pred_scaled)
        y_test_original = scaler.inverse_transform(y_test_scaled)

        # Evaluate the model
        mse = mean_squared_error(y_test, y_pred_scaled)
        mse_scores.append(mse)

        # Plot test predictions against original test data
        plt.figure(figsize=(15,6))
        plt.plot(y_test_original, label='True Values')
        plt.plot(y_pred, label='Predicted Values')
        plt.title(f'Test Predictions vs Original Test Data - Fold {i + 1}')
        plt.legend()
        plt.show()

    # Plot the training and validation loss after all folds
    plt.figure(figsize=(15,6))
    plt.plot(history.history['loss'], label='Overall Training Loss')
    plt.plot(history.history['val_loss'], label='Overall Validation Loss')
    plt.legend()
    plt.show()

    return mse_scores
```

```python
mse_scores = k_fold_cross_validation(X, y, scaler, k=5)
```

```python
print("Mean MSE:", np.mean(mse_scores))
print("Std MSE:", np.std(mse_scores))
```

```python
mse = np.mean(mse_scores)
```

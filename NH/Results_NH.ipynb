```python
scaler = MinMaxScaler()

def test_plots(df, bi_lstm, cnn_lstm, lstm, name):
    
    dates = df['Date'].tolist()
    datetime_index = pd.DatetimeIndex(dates)
    
    df['Extent'] = scaler.fit_transform(df['Extent'].values.reshape(-1, 1))

    extent = df['Extent']
    W_S=12
    X,y = fp.df_to_X_y(extent, W_S)
    
    bi_lstm_pred = bi_lstm.predict(X)
    cnn_lstm_pred = cnn_lstm.predict(X)
    lstm_pred = lstm.predict(X)
    
    bi_lstm_pred = scaler.inverse_transform(bi_lstm_pred)
    cnn_lstm_pred = scaler.inverse_transform(cnn_lstm_pred)
    lstm_pred = scaler.inverse_transform(lstm_pred)
    y = scaler.inverse_transform(y.reshape(-1,1))
     
    plt.figure(figsize=(12,6))
    plt.plot(datetime_index[12:], y, label='True Values')
    plt.plot(datetime_index[12:], bi_lstm_pred, label='BiDir')
    plt.plot(datetime_index[12:], cnn_lstm_pred, label='CNN+LSTM', color='red')
    plt.plot(datetime_index[12:], lstm_pred, label='LSTM')
    plt.title(f'{name} SIE')
    plt.legend()
    plt.show()
```

```python
test_plots(Baffin, baff_bi, baff_cnn, baff_lstm, 'Baffin_Bay')
test_plots(Barents, baren_bi, baren_cnn, baren_lstm, 'Barents_Sea')
test_plots(Bering, ber_bi, ber_cnn, ber_lstm, 'Bering_Sea')
test_plots(Greenland, green_bi, green_cnn, green_lstm, 'Greenland_Sea')
test_plots(Kara, kara_bi, kara_cnn, kara_lstm, 'Kara_Sea')
```

```python
scaler = MinMaxScaler()

def test_sub_plots(axs, df, bi_lstm, cnn_lstm, lstm):
    
    dates = df['Date'].tolist()
    datetime_index = pd.DatetimeIndex(dates)
    
    df['Extent'] = scaler.fit_transform(df['Extent'].values.reshape(-1, 1))

    extent = df['Extent']
    W_S=12
    X,y = fp.df_to_X_y(extent, W_S)
    
    bi_lstm_pred = bi_lstm.predict(X)
    cnn_lstm_pred = cnn_lstm.predict(X)
    lstm_pred = lstm.predict(X)
    
    bi_lstm_pred = scaler.inverse_transform(bi_lstm_pred)
    cnn_lstm_pred = scaler.inverse_transform(cnn_lstm_pred)
    lstm_pred = scaler.inverse_transform(lstm_pred)
    y = scaler.inverse_transform(y.reshape(-1,1))
     
    axs.plot(datetime_index[12:], y, label='True Values')
    axs.plot(datetime_index[12:], bi_lstm_pred, label='BiDir')
    axs.plot(datetime_index[12:], cnn_lstm_pred, label='CNN+LSTM', color='red')
    axs.plot(datetime_index[12:], lstm_pred, label='LSTM')
    axs.legend()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(16, 10))

test_sub_plots(axs[0,0], Baffin, baff_bi, baff_cnn, baff_lstm)
test_sub_plots(axs[0,1], Barents, baren_bi, baren_cnn, baren_lstm)
test_sub_plots(axs[1,0], Bering, ber_bi, ber_cnn, ber_lstm)
test_sub_plots(axs[1,1], Greenland, green_bi, green_cnn, green_lstm)
test_sub_plots(axs[2,0], Kara, kara_bi, kara_cnn, kara_lstm)

axs[0,0].set_title('Baffin Bay')
axs[0,1].set_title('Barents Sea')
axs[1,0].set_title('Bering Sea')
axs[1,1].set_title('Greenland Sea')
axs[2,0].set_title('Kara Sea')

axs[2,1].remove()

fig.suptitle('Actual vs Predicted')
for ax in axs.flat:
    ax.set(xlabel='Year', ylabel='Extent')

plt.tight_layout()

plt.show()
```

```python
def quad_plot(axs,df):
    coefficients = np.polyfit(df['Extent'].index, df['Extent'], 2)
    polynomial = np.poly1d(coefficients)
    
    axs.plot(df['Date'], df['Extent'], label='Sea Ice Extent', linestyle='-')
    axs.plot(df['Date'], polynomial(df['Extent'].index), label='Trend Line', linestyle='--', color='red')
    axs.set_xlabel('Date')
    axs.set_ylabel('Sea Ice Extent')
    axs.legend()
    axs.grid(True)
```

```python
fig, axs = plt.subplots(6, 2,  figsize=(15, 23))

quad_plot(axs[0, 0], Kara)
quad_plot(axs[0, 1], Barents)
quad_plot(axs[1, 0], Baffin)
quad_plot(axs[1, 1], Greenland)
quad_plot(axs[2, 0], Bering)
quad_plot(axs[2, 1], Beaufort)
quad_plot(axs[3, 0], Canadian)
quad_plot(axs[3, 1], Central_Arctic)
quad_plot(axs[4, 0], Chukchi)
quad_plot(axs[4, 1], East_Siberia)
quad_plot(axs[5, 0], Hudson)
quad_plot(axs[5, 1], Laptev)

axs[0,0].set_title('Kara Sea')
axs[0,1].set_title('Barents Sea')
axs[1,0].set_title('Baffin Bay')
axs[1,1].set_title('Greenland Sea')
axs[2,0].set_title('Bering Sea')
axs[2,1].set_title('Beaufort Sea')
axs[3,0].set_title('Canadian Archipelago')
axs[3,1].set_title('Central Arctic')
axs[4,0].set_title('Chukchi Sea')
axs[4,1].set_title('East Siberian Sea')
axs[5,0].set_title('Hudson Bay')
axs[5,1].set_title('Laptev Sea')

fig.suptitle('Quadratic Trends')
for ax in axs.flat:
    ax.set(xlabel='Year', ylabel='Extent')

plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

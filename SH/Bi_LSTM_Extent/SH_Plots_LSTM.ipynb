```python
df1 = Bell_Amundsen_Extent[Bell_Amundsen_Extent['Date'].dt.year == 2022]
df2 = Indian_Extent[Indian_Extent['Date'].dt.year == 2022]
df3 = Pacific_Extent[Pacific_Extent['Date'].dt.year == 2022]
df4 = Ross_Extent[Ross_Extent['Date'].dt.year == 2022]
df5 = Weddell_Extent[Weddell_Extent['Date'].dt.year == 2022]
```

```python
dfs = [df1, df2, df3, df4, df5]

combined_df = pd.DataFrame()

area_names = ['Bell_Amundsen', 'Indian', 'Pacific', 'Ross', 'Weddell']
```

```python
for i, (df, area_name) in enumerate(zip(dfs, area_names), start=1):
    df['Area'] = area_name
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df
```

```python
pivot_df = combined_df.pivot(index='Date', columns='Area', values='Extent')
```

```python
pivot_df
```

```python
pivot_df.plot(kind='area', stacked=True, figsize=(15, 6), colormap='cividis')

plt.title('Extent Values for Different Areas in 2022')
plt.xlabel('Date')
plt.ylabel('Extent Values')

plt.show()
```

```python
scaler = MinMaxScaler()

def test_plots(df, bi_lstm, cnn_lstm, lstm, name):
    
    df_original = df['Extent']
    previous_dates = df['Date'].tolist()
    
    dates = df['Date'][184:].tolist()
    datetime_index = pd.DatetimeIndex(dates)
    
    df['Extent'] = scaler.fit_transform(df['Extent'].values.reshape(-1, 1))

    extent = df['Extent']
    W_S=12
    X,y = fp.df_to_X_y(extent, W_S)
    X.shape, y.shape

    X_test, y_test = X[172:], y[172:]
    
    bi_lstm_pred = bi_lstm.predict(X_test)
    cnn_lstm_pred = cnn_lstm.predict(X_test)
    lstm_pred = lstm.predict(X_test)
    
    bi_lstm_pred = scaler.inverse_transform(bi_lstm_pred)
    cnn_lstm_pred = scaler.inverse_transform(cnn_lstm_pred)
    lstm_pred = scaler.inverse_transform(lstm_pred)
    
    plt.figure(figsize=(20,6))
    plt.plot(previous_dates, df_original, label='True Values')
    plt.plot(datetime_index, bi_lstm_pred, label='BiDir')
    plt.plot(datetime_index, cnn_lstm_pred, label='CNN+LSTM', color='red')
    plt.plot(datetime_index, lstm_pred, label='LSTM')
    plt.title(f'{name} SIE')
    plt.legend()
    plt.show()
```

```python
test_plots(Bell_Amundsen, ba_bi, ba_cnn, ba_lstm, 'Bell_Amundsen_Sea')
test_plots(Indian, ind_bi, ind_cnn, ind_lstm, 'Indian_Ocean')
test_plots(Pacific, pac_bi, pac_cnn, pac_lstm, 'Pacific_Ocean')
test_plots(Ross, ross_bi, ross_cnn, ross_lstm, 'Ross_Sea')
test_plots(Weddell, wedd_bi, wedd_cnn, wedd_lstm, 'Weddell_Sea')
```

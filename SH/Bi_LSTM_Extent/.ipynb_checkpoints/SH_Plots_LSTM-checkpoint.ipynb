```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
filepath = 'C:/Users/ss230/Documents/Study/NSIDC/Edited Data/S_Sea_Ice_Index_Regional_Monthly_Data_G02135_v3.0.xlsx'

Bell_Amundsen_Extent = pd.read_excel(filepath, sheet_name='Bell-Amundsen-Extent-km^2')
Indian_Extent = pd.read_excel(filepath, sheet_name='Indian-Extent-km^2')
Pacific_Extent = pd.read_excel(filepath, sheet_name='Pacific-Extent-km^2')
Ross_Extent = pd.read_excel(filepath, sheet_name='Ross-Extent-km^2')
Weddell_Extent = pd.read_excel(filepath, sheet_name='Weddell-Extent-km^2')
```

```python
import Function as fp
```

```python
fp.month_mapping(Bell_Amundsen_Extent)
fp.date_time(Bell_Amundsen_Extent)

fp.month_mapping(Indian_Extent)
fp.date_time(Indian_Extent)

fp.month_mapping(Pacific_Extent)
fp.date_time(Pacific_Extent)

fp.month_mapping(Ross_Extent)
fp.date_time(Ross_Extent)

fp.month_mapping(Weddell_Extent)
fp.date_time(Weddell_Extent)
```

```python
df1 = Bell_Amundsen_Extent[Bell_Amundsen_Extent['Date'].dt.year == 2022]
df2 = Indian_Extent[Indian_Extent['Date'].dt.year == 2022]
df3 = Pacific_Extent[Pacific_Extent['Date'].dt.year == 2022]
df4 = Ross_Extent[Ross_Extent['Date'].dt.year == 2022]
df5 = Weddell_Extent[Weddell_Extent['Date'].dt.year == 2022]
```

```python
dfs = [df1, df2, df3, df4, df5]

combined_df = pd.DataFrame()

area_names = ['Bell_Amundsen', 'Indian', 'Pacific', 'Ross', 'Weddell']
```

```python
for i, (df, area_name) in enumerate(zip(dfs, area_names), start=1):
    df['Area'] = area_name
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df
```

```python
pivot_df = combined_df.pivot(index='Date', columns='Area', values='Extent')
```

```python
pivot_df
```

```python
pivot_df.plot(kind='area', stacked=True, figsize=(15, 6), colormap='cividis')

plt.title('Extent Values for Different Areas')
plt.xlabel('Date')
plt.ylabel('Extent Values')

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df1['Date'], df1['Extent'], label='Bell_Amundsen')
plt.plot(df2['Date'], df2['Extent'], label='Indian')
plt.plot(df3['Date'], df3['Extent'], label='Pacific')
plt.plot(df4['Date'], df4['Extent'], label='Ross')
plt.plot(df5['Date'], df5['Extent'], label='Weddell')

plt.xlabel('Date')    
plt.ylabel('Extent')
plt.title('Extent Values in 2022')
plt.legend()

plt.show()
```
